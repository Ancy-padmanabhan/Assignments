```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('/content/tip-amount.xlsx')
```

```python
data.head()
```

```python
corrmatrix=data.corr()
plt.subplots(figsize=(10,4))
sns.heatmap(corrmatrix,annot=True)
```

```python
data.columns
```

```python
y=data['Observed tip amount(yi)']
```

```python
X=data.drop(['Observed tip amount(yi)'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=20, test_size=0.20)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
model = lr.fit ( X_train , y_train )
```

```python
predictions=model.predict(X_test)
```

```python
X_test
```

```python
predictions
```

```python
new_data_points=pd.DataFrame({'Meal':[6,10],'Observed total bill(xi)':[130,180]})
predicted_tip_amount=model.predict(new_data_points)
```

```python
predicted_tip_amount
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('\n MSE is:',mean_squared_error(y_test,predictions))
```

```python
from sklearn.metrics import r2_score
print('The r squared value is:',r2_score(y_test,predictions))
```
