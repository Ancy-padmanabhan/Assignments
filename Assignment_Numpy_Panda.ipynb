## 5.**Perform meaningful visualizations using the dataset**

**SL v/s PL**

```python
data.plot(kind='scatter', x='SL', y='PL')
plt.show()
```

**Line Chart showing the Sepal length measured in the dataset to find the maximum value captured**

```python
data['SL'].plot(kind='line', figsize=(8, 4), title='SL')
```

**The species/classifications of the data set and the count**

```python
data.groupby('Classification').size().plot(kind='barh', color=sns.palettes.mpl_palette('tab20'))
```

**Sepal Length and width Comparisons as scatter plot**

```python
data.plot(kind='scatter', x='SL', y='SW',s=32, alpha=.8,color='green')
```

**Variables measuring properties of same entity. Sepal Length and Sepal width. Petal length and petal width are plotted here to see if there is any relation between the same to identify the classifications**

```python
sns.jointplot(data,x='SL',y='SW',kind='kde',color="Purple")
sns.jointplot(data,x='PL',y='PW',kind='kde')
plt.savefig("kde.png")
plt.show()
```

## **Sepal Length Density Classification wise comparison**

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(15,5), sharex=False)
sns.distplot(data['SL'],fit=stats.expon,hist=True,kde=True,ax=ax1)
sns.violinplot(data, x='Classification',y='SL',inner='box',ax=ax2)
```
