```python
data=pd.read_csv('beer-servings.csv',index_col='Unnamed: 0')
data
```

```python
for i in ['beer_servings','wine_servings','spirit_servings','total_litres_of_pure_alcohol']:
  data[i]=data[i].fillna(data[i].median())
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label_en=LabelEncoder()
```

```python
data.nunique()
```

```python
data1=data.copy()
data1.isna().sum()
```

```python
data1['continent']=label_en.fit_transform(data1['continent'])
data1.head()
```

```python
# One Hot Encoder
data2=data.copy()
```

```python
data2=data2.drop('country', axis = 1)
```

```python
data2=pd.get_dummies(data2,dtype=int)
data2.head()
```
