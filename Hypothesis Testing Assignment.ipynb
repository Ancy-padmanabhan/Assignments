**A company started to invest in digital marketing as a new way of their product promotions. For that they collected data and decided to carry out a study on it.**

```python
#1.The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.
```

```python
import numpy as np
import pandas as pd
import statistics as stat
import statsmodels.api as sm
from statsmodels.stats.weightstats import ztest
```

```python
data=pd.read_csv('Sales_add.csv')
data.head()
```

```python
data.info()
```

```python
z_stat,p_value=ztest(data['Sales_before_digital_add(in $)'],data['Sales_before_digital_add(in $)'])
```

```python
if p_value<0.05:
  print("reject null Hypothesis")
else:
  print("failed to reject null hypothesis")
```

**according to the ztest results There is a surge in sales after stepping into digital marketing**

```python
#2.The company needs to check whether there is any dependency between the features “Region” and “Manager”.
from scipy.stats import chisquare
from scipy.stats import chi2_contingency
```

```python
contingency=pd.crosstab(data['Region'],data['Manager'])
contingency
```

```python
chi2stat,p_value,dof,exp=chi2_contingency(contingency)
p_value
```

```python
if p_value<0.05:
  print("reject null Hypothesis")
else:
  print("failed to reject null hypothesis")
```

**As per the chisquare test on the region and the manager there is a dependency.**
